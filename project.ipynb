```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


cap = cv2.VideoCapture('v/v.mp4')
count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
bg = cv2.imread('v/bg.png')
bgg = cv2.cvtColor(bg, cv2.COLOR_BGR2GRAY)
frames = []
for i in range(count):
    _, img = cap.read()
    temp = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    frames.append(temp)

for i in range(len(bgg)):
    for j in range(len(bgg[0])):
        values = np.array([ele[i][j] for ele in frames])
        median = np.median(values)
        bgg[i][j] = int(median)

plt.imshow(bgg, cmap = plt.cm.gray)
plt.show()
```

```python
import cv2

orb = cv2.xfeatures2d.SIFT_create()
t1 = cv2.imread('frame1.png', 0)
t2 = cv2.imread('frame2.png', 0)
# t1 = t1[100:180, 35:170]
# t2 = t2[100:180, 140:270]
print(type(t1))
print(type(t2))
# t1 = res1
# t2 = res2
# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(t1,None)
kp2, des2 = orb.detectAndCompute(t2,None)

bf = cv2.BFMatcher(cv2.NORM_L1,crossCheck=False)
matches = bf.match(des1,des2)

# BFMatcher with default params
# bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# # Match descriptors.
# matches = bf.match(des1,des2)

# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# Draw first 10 matches.
img3 = cv2.drawMatches(t1,kp1,t2,kp2,matches[:30], None, flags=2)

plt.imshow(img3)
plt.show()
```

## Calculate the Motion Vector

```python
print(kp1[0].pt)
print(kp2[0].pt)
print(type(list(kp1[0].pt)))
np.array(list(kp1[0].pt))

numOfMatches = len(kp1)
shift = []
for i in range(numOfMatches):
    vec = np.array(list(kp2[i].pt)) - np.array(list(kp1[i].pt))
    shift.append(vec)
    
shift = np.array(shift)
motion = np.mean(shift, axis=0)
print(motion)
```

## Retrieve the Object Segmentation

```python
img1 = cv2.imread('frame1.png')              # img.shape : (413, 620, 3)
mask1 = np.zeros(img1.shape[:2],np.uint8)   # img.shape[:2] = (413, 620)
# img2 = cv2.imread('frame2.png')              # img.shape : (413, 620, 3)
# mask2 = np.zeros(img2.shape[:2],np.uint8)

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (30,110,140,60)

# this modifies mask 
cv2.grabCut(img1,mask1,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)

# If mask==2 or mask== 1, mask2 get 0, other wise it gets 1 as 'uint8' type.
mask2 = np.where((mask1==2)|(mask1==0),0,1).astype('uint8')

# adding additional dimension for rgb to the mask, by default it gets 1
# multiply it with input image to get the segmented image
img_cut = img1*mask2[:,:,np.newaxis]

plt.subplot(211),plt.imshow(img1)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(212),plt.imshow(img_cut)
plt.title('Grab cut'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
print(img1.shape)
print(mask2.shape)
step = 10
alpha = 0.8
unitMotion = motion / np.sqrt(motion[0]**2 + motion[1]**2) * step

unitMotion = np.uint8(unitMotion)
print(unitMotion)
result = np.zeros(shape = img1.shape)
for i in range(len(img1)):
    for j in range(len(img1[0])):
        if i+unitMotion[1] < len(img1) and j+unitMotion[0] < len(img1[0]) and mask2[i + unitMotion[1]][j + unitMotion[0]] == 1:
#             result[i][j] = img1[i][j] * img1[i + unitMotion[1]][j + unitMotion[0]]
            result[i][j] = cv2.addWeighted(img1[i][j], alpha, img1[i + unitMotion[1]][j + unitMotion[0]], 1-alpha, 0.0).flatten()
        else:
            result[i][j] = img1[i][j]
            
result = np.uint8(result)
plt.imshow(result)
plt.show()
```
